# Exploratory data analysis

## Summary of datasets

There are nine datasets in total, summarised as follows:

- **'train.csv'** contains the hourly (normalised) wind power measurements from all seven windfarms.  
- **'windforecasts_wf1.csv'** contains 48 hour ahead hourly forecasts for wind speed and direction, where forecasts are produced at 12 hour intervals, for wind farm 1. There are six other such datasets corresponding to wind forecasts for the other wind farms.  
- **'benchmark.csv'** provides example forecast results from the persistence forecast method (i.e. the forecast is that the power output doesn't change), which  acts as a template for submission to kaggle.

The power output data is designed so that the period between 2009/7/1 and 2010/12/31 is to be used for model identification and training, and the remainder of the data is the hold-out set for testing on kaggle. The hold-out set is designed to test 48 hour ahead forecasts for hourly power output, and so has a repeating pattern of 48 hours of missing data, followed by 36 hours of present data. 

The code and graphs below check that this pattern exists in the data.

```python
# look at gaps in the time series data
no_gaps_df = train.resample('H').asfreq() # resample so no hours are missing (will be filled with NaN)
gaps = no_gaps_df.isna().astype(int) # create boolean for gap or no gap

# plot gaps
fig, ax = plt.subplots(7, 1, figsize=(15, 7))
for i in range(7):
    column = 'wp'+str(i+1)
    ax[i].plot(gaps.index, gaps[column], lw=1)
    ax[i].set_title('Power ouput - wind farm '+str(i+1))
plt.tight_layout()
```

```python
# set up train, test and hold-out sets

# the first data gap is on 01/01/2011
# split 36 hours prior to this to include the available power data before the first forecasting period 
ext_split_date = datetime(2011,1,1,1) - pd.Timedelta(hours = 36) 

# use 16 weeks of testing data (~20% of full training set)
int_split_date = ext_split_date - pd.Timedelta(weeks = 16) 

power_train = train.loc[:int_split_date - pd.Timedelta(hours = 1), :] # datetimes use inclusive indexing
power_test_int = train.loc[int_split_date:ext_split_date - pd.Timedelta(hours = 1),:]
power_test_ext = train.loc[ext_split_date:,:]

# find number of repeated patterns expected in hold-out set - note we expect two per week
num_days = power_test_ext.index[-1] - power_test_int.index[-1]
num_patterns = num_days/pd.Timedelta(days=3, hours=12)

# write expected pattern to list - repeating pattern of 36 hour period of data, then 48 period without data
# however, the last 48 period of missing data is not included in the test index, so this is accounted for
expected_pattern = ([0]*36+[1]*48)*int(np.floor(num_patterns)) + [0]*36

# test number of unexpected gaps in hold out test set
ext_test_gaps = gaps.loc[ext_split_date:,:]
print('Number of unexpected gaps:',(expected_pattern != ext_test_gaps).sum().sum())
```

```python
# save datasets
power_train.to_csv('../datasets/power_train.csv')
power_test_int.to_csv('../datasets/power_test_int.csv')
power_test_ext.to_csv('../datasets/power_test_ext.csv')
```
