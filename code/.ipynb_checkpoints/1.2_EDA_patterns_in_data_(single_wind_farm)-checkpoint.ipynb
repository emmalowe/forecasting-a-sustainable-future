# Exploratory data analysis - patterns in the data (single wind farm)

Before creating a forecasting model, it is desirable to remove trends and seasonality in the data to create a stationary time series. This notebook explores patterns in the data, as preparation for modeling. The baseline RMSE from the persistent forecasting method is also calculated.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# read in data

power_train = pd.read_csv('../datasets/power_train.csv', index_col=0)
power_train.index = pd.DatetimeIndex(power_train.index, freq='H')

power_val_target = pd.read_csv('../datasets/power_val_target.csv', index_col=0)
power_val_target.index = pd.DatetimeIndex(power_val_target.index)

power_val_benchmark = pd.read_csv('../datasets/power_val_benchmark.csv', index_col=0)
power_val_benchmark.index = pd.DatetimeIndex(power_val_benchmark.index)

# restrict analysis to one wind farm

wp1_train = power_train['wp1']
wp1_val_target = power_val_target['wp1']
wp1_val_benchmark = power_val_benchmark['wp1']
```

## Patterns in the power data

```python
# plot data
plt.figure(figsize = (20,3))
plt.plot(wp1_train.index, wp1_train.values);
```

```python
# difficult to see patterns, so split into intervals for plotting

num_intervals = 4
len_interval = int(np.ceil(len(wp1_train)/4))

fig, ax = plt.subplots(num_intervals,1,figsize = (20, 3*num_intervals))
for i in range(num_intervals):
    ax[i].plot(wp1_train.index[i*len_interval:(i+1)*len_interval], wp1_train.values[i*len_interval:(i+1)*len_interval])

ax[0].set_title('Hourly power output from a single wind farm over 62 weeks\n', fontsize = 24);
```

```python
# no discernible pattern - check seasonal decomposition
```

```python
decomp = seasonal_decompose(wp1_train)
decomp.plot();
```

```python
# seasonal pattern is insignificant and trend is not smooth enough!
```

```python
# check for stationarity

# Dickey-Fuller test:
print('Results of Dickey-Fuller Test:')
dftest = adfuller(wp1_train)
dfoutput = pd.Series(dftest[0:2], index=['Test Statistic','p-value'])
print(dfoutput)
```

```python
# reject the null hypothesis that timeseries has a unit-root
# accept the alternative hypothesis that timeseries is stationary
# therefore can proceed to modeling
```

## Baseline RMSE

```python
# calculate baseline RMSE to provide a target for models
base_RMSE = mean_squared_error(y_true = wp1_val_target, y_pred = wp1_val_benchmark)**0.5
print('Baseline RMSE:', base_RMSE)
```
