# Modeling - Time series (power only)

This notebook focuses on stage 1 of modeling for this project: modeling power output at one wind farm, using a time series model that is built purely on previous power output.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import statsmodels.api as sm  
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARMA
from statsmodels.tsa.statespace import sarimax
from statsmodels.tsa.stattools import adfuller
```

```python
from sklearn.metrics import mean_squared_error
```

```python
power_train = pd.read_csv('../datasets/power_train.csv', index_col=0)
power_train.index = pd.DatetimeIndex(power_train.index, freq='H')
```

## Patterns in the data

```python
# restrict analysis to one wind farm
wp1_train = power_train['wp1']
```

```python
# re-centre data around 0, with range (-1,1)
wp1_train = (wp1_train)*2 - 1
```

```python
# plot data
plt.figure(figsize = (20,3))
plt.plot(wp1_train.index, wp1_train.values)
plt.axhline(y=0, c='k', lw=1)
```

```python
# difficult to see patterns, so split into intervals for plotting

num_intervals = 4
len_interval = int(np.ceil(len(wp1_train)/4))

plt.subplots(num_intervals,1,figsize = (20, 3*num_intervals))
for i in range(num_intervals):
    plt.subplot(num_intervals,1,i+1)
    plt.plot(wp1_train.index[i*len_interval:(i+1)*len_interval], wp1_train.values[i*len_interval:(i+1)*len_interval])
    plt.axhline(y=0, c='k', lw=1)
```

```python
# no discernible pattern - check seasonal decomposition
```

```python
decomp = seasonal_decompose(wp1_train)
decomp.plot();
```

```python
# seasonal pattern is insignificant and trend is not smooth enough!
```

```python
# check for stationarity

# Dickey-Fuller test:
print('Results of Dickey-Fuller Test:')
dftest = adfuller(wp1_train)
dfoutput = pd.Series(dftest[0:2], index=['Test Statistic','p-value'])
print(dfoutput)
```

```python
# reject the null hypothesis that timeseries has a unit-root
# accept the alternative hypothesis that timeseries is stationary
# therefore can proceed to modeling
```

## Auto-correlation

```python
# look at ACF and PACF plots
```

```python
plot_acf(wp1_train, lags=100);
```

```python
plot_pacf(wp1_train, lags=100);
```

```python
# PACF has a sharp cut-off after 2 lags, and the lag-1 ACF is positive
# this signature could be suggestive of a AR(2) model
```

## Fit model

```python
p = 2 # num of autoregressive terms
q = 0 # num of MA terms

# use SARIMAX for additional functionality later on
ar2 = sarimax.SARIMAX(endog = wp1_train, order = (p,0,q)).fit()
ar2.summary()
```

## Test model

The test set needs to be designed in the same way as the hold-out test set. That is, set up in such a way that following a 36 hour period where power output is available, there is a 48 hour forecasting period where there is no data. In this period, hourly forecasts are made, which form the model predicitions. 

#### Set up data and target datasets and internal benchmark (based on the persistant forecast method)

```python
power_test_int = pd.read_csv('../datasets/power_test_int.csv', index_col = 0)
power_test_int.index = pd.to_datetime(power_test_int.index)

# find number of repeated patterns (36 hours of data, 48 hours of missing data)
num_patterns = int(power_test_int.shape[0]/84)
```

```python
# create index list for data and target datasets

data_index = []
target_index = []

for pattern_num in range(num_patterns):
    data_index.extend(list(range(84*pattern_num, 84*pattern_num+36)))
    target_index.extend(list(range(84*pattern_num+36, 84*pattern_num+84)))
```

```python
# create data and target dataframes
power_test_int_data = power_test_int.iloc[data_index,:]
power_test_int_target = power_test_int.iloc[target_index,:]
```

```python
# create benchmark model based on persistant forecast method

# resample data using forward filling to get persistant forecast
power_test_int_benchmark = power_test_int_data.reindex(power_test_int.index, method='ffill')
# reduce to just forecasted data
power_test_int_benchmark = power_test_int_benchmark.iloc[target_index,:]
```

```python
power_test_int_data.to_csv('../datasets/power_test_int_data.csv')
power_test_int_target.to_csv('../datasets/power_test_int_target.csv')
power_test_int_benchmark.to_csv('../datasets/power_test_int_benchmark.csv')
```

#### Baseline RMSE

```python
# wp1 only
base_RMSE = mean_squared_error(y_true = power_test_int_target['wp1'], y_pred = power_test_int_benchmark['wp1'])**0.5
print('Baseline RMSE:', base_RMSE)
```

#### Testing AR(2) model

```python
# model predictions

# for each 36 hour period, read in data and make 48 hour forecast
predictions = []
for pattern_num in range(num_patterns):
    data = power_test_int_data.iloc[36*pattern_num:36*(pattern_num+1),0] # looking at first wind farm only
    # reset index frequency to hourly - this was lost before when slicing up the date-time index
    # need this so the model knows who frequently to make predictions
    data.index = pd.DatetimeIndex(data.index, freq='H')
    # reinstantiate model, but use old parameters
    model = sarimax.SARIMAX(endog = data, order = (2,0,0))
    model = model.filter(ar2.params)
    # forecast
    predictions.extend(model.forecast(48))
    
ar2_predictions = pd.DataFrame(predictions, columns=['wp1'], index = power_test_int_target.index)
```

```python
ar2_RMSE = mean_squared_error(y_true = power_test_int_target['wp1'], y_pred = ar2_predictions['wp1'])**0.5
print('AR2 model RMSE:', ar2_RMSE)
```
